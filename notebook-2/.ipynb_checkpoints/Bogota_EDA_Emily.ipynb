```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("translated_bogota.csv")
df
```

```python
df_1 = df.loc[df['Age Group'] == "Boys, girls and adolescents or minors (<18 years)", :].groupby("Year").mean().reset_index()
df_1
```

```python
plt.plot(df_1["Year"], df_1["Men (rate per hundred thousand inhabitants)"], label = "Men");
plt.plot(df_1["Year"], df_1["Women (rate per hundred thousand inhabitants)"], label = "Women");
plt.legend();
plt.ylabel("Rate per hundred thousand inhabitants");
plt.show();
```

```python
df_2 = df.loc[df['Age Group'] == "Adults or of legal age (> 18 years)", :].groupby("Year").mean().reset_index()
df_2
```

```python
plt.plot(df_2["Year"], df_2["Men (rate per hundred thousand inhabitants)"], label = "Men");
plt.plot(df_2["Year"], df_2["Women (rate per hundred thousand inhabitants)"], label = "Women");
plt.legend();
plt.ylabel("Rate per hundred thousand inhabitants");
plt.show();
```

```python
df_3 = df.groupby(["Year", "Age Group"]).mean().reset_index()
df_3
```

```python
sns.barplot(x="Year", y="Men (rate per hundred thousand inhabitants)", hue="Age Group", data = df_3);
```

```python
sns.barplot(x="Year", y="Women (rate per hundred thousand inhabitants)", hue="Age Group", data = df_3);
```

```python
df_4 = df.loc[df['Age Group'] == "Boys, girls and adolescents or minors (<18 years)", :].groupby("Fact department").mean().reset_index()
df_4
```

```python
sns.barplot(x="Fact department", y="Men (rate per hundred thousand inhabitants)", data = df_4);
plt.xticks(rotation=90);
```

```python
sns.barplot(x="Fact department", y="Women (rate per hundred thousand inhabitants)", data = df_4);
plt.xticks(rotation=90);
```
